COMMON EXAMPLES OF SPECIAL METHODS

1. Constructor: _init_()
Think of this as the "factory settings" or the setup wizard for your object. The moment you create something new from a class, this method runs automatically to make sure it starts with the right properties (like giving a new user a name and an email right away).It initializes a newly created instance of a class. It is automatically called when an object is instantiated and allows you to set the initial state (attributes) of the object.

```python
class Book:
    def __init__(self, title, author):
        self.title = title       # Initializing the title attribute
        self.author = author     # Initializing the author attribute

# Creating (instantiating) an object
my_book = Book("Python Basics", "Alice")
print(my_book.title)
```

2. String Representation: _str() and __repr_()
_str_() is the friendly, readable description meant for everyday people to look at (like a product label).
_repr_() is the detailed, official description meant for developers or debugging (like the product's internal serial number and specs).
_str_ returns an informal, user-friendly string representation of an object. _repr_ returns an official, unambiguous string representation used for debugging and logging, which ideally looks like a valid Python expression to recreate the object.

```python
class Car:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    def __str__(self):
        return f"{self.brand} {self.model}"

    def __repr__(self):
        return f"Car(brand='{self.brand}', model='{self.model}')"

my_car = Car("Toyota", "Corolla")
print(str(my_car))   # Triggers __str__
print(repr(my_car))  # Triggers __repr__
```

3. Length: __len__()
This allows you to use Python's built-in len() tool on your own custom objects. It tells Python exactly what to count when someone asks how long your object is. It defines the behavior of the built-in len() function for an instance of your class. It must return a non-negative integer representing the size or number of elements in the object.

```python
class Team:
    def _init_(self, members):
        self.members = members  # Expecting a list of names

    def _len_(self):
        return len(self.members)

dev_team = Team(["Alice", "Bob", "Charlie"])
print(len(dev_team))
```

4. Equality: __eq__()
By default, Python checks if two objects are the exact same physical piece of data in computer memory. This method lets you change the rules so that two different objects are considered "equal" if their internal details match (like two identical, separate copies of the same book).It overrides the equality operator (==). It defines how instances of a class are compared for structural equality rather than identity (memory location).

```python
class Product:
    def _init_(self, id, name):
        self.id = id
        self.name = name

    def _eq_(self, other):
        if isinstance(other, Product):
            return self.id == other.id  # Equal if their IDs match
        return False

p1 = Product(101, "Laptop")
p2 = Product(101, "Laptop")
print(p1 == p2)
```

5. Operator Overloading: __add__() and __sub__()
This teaches Python how to use math symbols like plus (+) and minus (-) on your custom objects, which normally wouldn't make sense to a computer (like adding two shopping carts together).
It implements operator overloading for addition (+) and subtraction (-). It defines the custom behavior of arithmetic operators when applied to class instances.

```python
class Wallet:
    def _init_(self, cash):
        self.cash = cash

    def _add_(self, other):
        return Wallet(self.cash + other.cash)

    def _sub_(self, other):
        return Wallet(self.cash - other.cash)

w1 = Wallet(50)
w2 = Wallet(20)
combined_wallet = w1 + w2
print(combined_wallet.cash)
```

6. Item Access: __getitem__()
This lets you use square brackets [ ] on your object to grab pieces of data inside it using an index or a key, just like you would pull an item out of a Python list or dictionary.
It implements evaluation of the routing/indexing operator (self[key]). It allows instances of a class to use indexing or key-based lookups.

```python
class Playlist:
    def _init_(self, songs):
        self.songs = songs

    def _getitem_(self, index):
        return self.songs[index]

my_playlist = Playlist(["Song A", "Song B", "Song C"])
print(my_playlist[1])  # Accesses the second item
```

7. Item Assignment: __setitem__()
This works alongside item access, letting you use square brackets [ ] to change or assign a value inside your object (like swapping out a song at a specific position in your playlist). It implements assignment to an indexed element (self[key] = value). It allows you to mutate or assign values to specific keys or indices within an object.

```python
class Playlist:
    def __init__(self, songs):
        self.songs = songs

    def __setitem__(self, index, value):
        self.songs[index] = value

my_playlist = Playlist(["Song A", "Song B", "Song C"])
my_playlist[0] = "New Song"  # Changes "Song A" to "New Song"
print(my_playlist.songs)
```

8. Membership Testing: __contains__()
This allows you to use the in keyword to check if an item exists inside your object (like asking Python: "Is chocolate in this grocery list?"). It implements the membership test operators (in and not in). It should return a boolean (True or False) based on whether an item exists within the object's collection.

```python
class Inventory:
    def _init_(self, items):
        self.items = items

    def _contains_(self, item):
        return item in self.items

my_stock = Inventory(["apple", "banana", "orange"])
print("apple" in my_stock)
```

9. Callable Objects: _call_()
This turns your object into something you can treat like a regular function. You can run or "call" the object itself using parentheses () just like a function action. It allows an instance of a class to be called as a function. When an object is invoked with parentheses, this method is executed.

```python
class Greeter:
    def _init_(self, greeting):
        self.greeting = greeting

    def _call_(self, name):
        return f"{self.greeting}, {name}!"

say_hello = Greeter("Hello")
# Treating the object like a function
print(say_hello("Alice"))
```

10. Iteration Support: __iter__() and __next__()
This makes your object loop-friendly. It allows your custom object to work smoothly inside a standard for loop, serving up items one by one until it runs out. Together, these realize the Python Iterator Protocol. __iter__ returns the iterator object itself. __next__ returns the next item in the sequence and raises a StopIteration exception when no more elements are available.

```python
class Countdown:
    def _init_(self, start):
        self.current = start

    def _iter_(self):
        return self

    def _next_(self):
        if self.current <= 0:
            raise StopIteration
        self.current -= 1
        return self.current + 1

# Using the object directly in a loop
for num in Countdown(3):
    print(num)
```

11. Boolean Evaluation: _bool_()
This lets you decide when your custom object should count as "True" or "False" when used in an if statement or truth test. It implements truth value testing for an object. Used by Python's built-in bool() function and conditional logic; it must return either True or False.

```python
class Account:
    def _init_(self, balance):
        self.balance = balance

    def _bool_(self):
        return self.balance > 0  # True if balance is positive, otherwise False

my_account = Account(100)
if my_account:
    print("Account is active and funded!")
```

12. Destructor: _del_()
This is the clean-up crew. It runs automatically right before the computer completely deletes the object from its memory, giving it a chance to close files or clean up mess. It is the destructor method. It is called when an object is about to be destroyed or garbage collected, allowing for final resource cleanup (like closing database connections).

```python
class DatabaseConnection:
    def _init_(self):
        print("Connection opened.")

    def _del_(self):
        print("Connection closed safely.")

# Creating the object
conn = DatabaseConnection()
# Explicitly deleting the object to trigger _del_
del conn
```
